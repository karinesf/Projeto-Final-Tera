```python
#Importar as bibliotecas necessárias para o webscrapping
!pip install beautifulsoup4
!pip install pandas
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import csv
```

```python
#Criar a lista de links dos jogos do torneio
href=[]
list_df_scoreboards = []
end = "https://ph-mpl.com/"  #Endereço das páginas com as lista dos anuncios
url = urlopen(end)
soup = BeautifulSoup(url.read(),"html.parser") #Leio o html
#Achar as classes do html onde estão os links
a = soup.find_all("a",{"class":"button button-mini button-match-data"})
for x in a:    
    href.append(x['href'])
print(href[27])
print(href[58])
#Deletar os jogos que não possuem dados
del href[27]
del href[58]
```

```python
#Criar uma função para listar os dados do Summary
def getSummary(link):
    list_games = ["game1","game2","game3","game4","game5"]
    list_df_summarys = []
    for game in list_games:
        print(game)
        url_matches = urlopen(link)
        soup_matches = BeautifulSoup(url_matches.read(),"html.parser")
        div = soup_matches.find("div", {"id":game})
        #Achar os dados conforme a sua localização do html (se diferente de nenhum...)
        if(div != None):
            print('Teve esse jogo ...')
            h3 = div.find_all("h3", {"class":"mb-0 mt-1"})[0].string
            h4 = div.find_all("h5", {"class":"mb-0"})[0:34:3]
            h5 = div.find_all("h3", {"class":"mb-0 mt-1"})[1].string
            h6 = div.find_all("h5", {"class":"mb-0"})[2:37:3]
            summary = {
            "id":[link,link],
            "game":[game,game],
            "team_name": [h3[:4],h5[:4]],
            "total_kills": [h4[0].string.replace("\n", "").strip(),h6[0].string.replace("\n", "").strip()],
            "total_assists": [h4[1].string.replace("\n", "").strip(),h6[1].string.replace("\n", "").strip()],
            "total_deaths": [h4[2].string.replace("\n", "").strip(),h6[2].string.replace("\n", "").strip()],
            "total_gold": [h4[3].string.replace("\n", "").strip(),h6[3].string.replace("\n", "").strip()],
            "gold_min": [h4[4].string.replace("\n", "").strip(),h6[4].string.replace("\n", "").strip()],
            "total_damage": [h4[5].string.replace("\n", "").strip(),h6[5].string.replace("\n", "").strip()],
            "damage_min": [h4[6].string.replace("\n", "").strip(),h6[6].string.replace("\n", "").strip()],
            "red_buff": [h4[7].string.replace("\n", "").strip(),h6[7].string.replace("\n", "").strip()],
            "blue_buff": [h4[8].string.replace("\n", "").strip(),h6[8].string.replace("\n", "").strip()],
            "lord_kill": [h4[9].string.replace("\n", "").strip(),h6[9].string.replace("\n", "").strip()],
            "tortoise_kill": [h4[10].string.replace("\n", "").strip(),h6[10].string.replace("\n", "").strip()],
            "tower_destroy": [h4[11].string.replace("\n", "").strip(),h6[11].string.replace("\n", "").strip()],
            }
            #Transformar os dados em DataFrames e adicionando os jogos no mesmo csv por partida
            df_summary = pd.DataFrame(summary)
            list_df_summarys.append(df_summary)
            concat = pd.concat(list_df_summarys)
        else:
            print('Não teve esse jogo ...')
    return concat
```

```python
#Extrair os dados em um for loop, que emite um arquivo csv a partir dos links
list_summary = []
for link in href[57:]: #0-20, 20-40, 40-...
    print(link)
    result_summary = getSummary(link)
    csv_text = '.csv'
    name_csv = sub_link = link[30:] + csv_text
    print(name_csv)
    result_summary.to_csv(name_csv)
```
